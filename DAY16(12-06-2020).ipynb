### Linear Algebra

```python
from scipy import linalg
# create 2d Matrix
ar2 = np.array([[4,5],[3,2]])
linalg.det(ar2)
```

```python
linalg.inv(ar2)
```

## Image operations

```python
from scipy import misc
data = misc.face()
plt.imshow(data)
```

```python
flip_data = np.flipud(data)
plt.imshow(flip_data)
```

```python
rand_arr = np.random.randint(low=0,high=225,size=data.shape)
plt.imshow(rand_arr)
```

```python
from scipy import ndimage
data_rotate = ndimage.rotate(data,135)
plt.imshow(data_rotate)
```

```python
from scipy import ndimage
data_rotate = ndimage.rotate(data,45)
plt.imshow(data_rotate)
```
